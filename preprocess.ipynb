```python
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import csv
import pickle
import numpy as np
import pandas as pd
import pickle
import jdatetime
#starting date = 29 tir
```

```python
df = pd.read_csv("data.csv")
# Define the start date
start_date = jdatetime.date(1402, 4, 28) 
```

```python
#df.head()
```

```python
# Dictionary for the column mappings
column_dict = {
    'اندازه سازمان' : 'Company Size',
    'صنعت' : 'Industry',
    'جنسیت': 'Gender',
    'تحصیلات' : 'Education',
    'زبان' : 'Language',
    'نرم افزارها' : 'Software',
    'خدمت سربازی' : 'Military service',
    'سن' : 'Age'
}

# For each key-value pair in the dictionary
for persian_column, english_column in column_dict.items():

    df[persian_column].fillna(df[english_column], inplace=True)
```

```python
del df["Company Size"],df["Industry"],df["Gender"],df["Education"],df["Language"],df["Software"],df["Military service"],df["Age"]
```

```python
df.rename(columns={'زبان': 'language', 'روز و ساعت کاری': 'workdays','حقوق':"salary",
                   "جنسیت":"gender",'تحصیلات':'education','اندازه سازمان':"company_size",
                   'مکان':'location','مزایا و تسهیلات':"bonus",'صنعت':'industry',
                    'سفرهای کاری':'business_trip','military':'خدمت سربازی'},inplace=True)
```

```python
df = df.rename(columns={
    'زمان':'time',
    'سن' : 'age',
    'نام آگهی' : 'ad_name',
    'نرم افزارها': 'softwares',
    'خدمت سربازی':'military_services',
    'سابقه کاری': 'experience'
})
```

```python
df["time"] = df["time"].apply(lambda x: str(x).strip("()"))
df["time"] = df["time"].apply(lambda x: x.split(" ")[0])
```

```python
df["time"] = df["time"].apply(lambda x: 0 if x=="امروز" else x)
df["time"] = df["time"].apply(lambda x: 1 if x=="دیروز" else x)
df["time"] = df["time"].apply(lambda x: 30 if x=="nan" else x)
df["time"] = df["time"].apply(lambda x: 0 if x=="Today" else x)
df["time"] = df["time"].apply(lambda x: 1 if x=="Yesterday" else x)
```

```python

# Convert the 'time' column to a list of timedelta objects
time_deltas = [jdatetime.timedelta(days=x) for x in df['time'].astype("int")]

# Subtract each timedelta from the start date
df['date'] = [start_date - td for td in time_deltas]
del df["time"]
```
